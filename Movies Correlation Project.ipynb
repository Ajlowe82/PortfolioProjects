```python
# Import libraries

import pandas as pd
import seaborn as sb
import numpy as np
import re

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (17,8) #Adjusts the configuration of the plots we will create

# Read in the data

df = pd.read_csv(r'C:\Users\Alana\Downloads\movies.csv')
```

```python
# Let's lookat the data

df.head()
```

```python
# Check for missing data

for col in df.columns:
   pct_missing = np.mean(df[col].isnull())
   print('{} - {}%'.format(col, pct_missing)) 
```

```python
# Data type for columns

df.dtypes
```

```python
# Change data type of the budget, gross, votes columns

df['budget'] = df['budget'].fillna(0).astype('int64')

df['gross'] = df['gross'].fillna(0).astype('int64')

df['votes'] = df['votes'].fillna(0).astype('int64')
```

```python
df.head()
```

```python
# Create correct year column

df['yearcorrect'] = df['released'].astype(str).str.extract(',(.+?) \(')
df.head()
```

```python
# Sort by gross earnings

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Look at all data in dataframe

pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Prediction: Budget and company will have a high correlation

# Scatter plot

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')

plt.show()
```

```python
df.head()
```

```python
 # Plot budget vs gross using seaborn
    
sb.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Let's start looking at correlation

df.corr(method='pearson', numeric_only = True) #pearson, kendall, spearman
```

```python
# High correlation between budget and gross

correlation_matrix = df.corr(method='pearson', numeric_only = True)
sb.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```
